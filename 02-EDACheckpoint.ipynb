# Research Question

What is the significance of the Yen carry trade as it relates to global markets?
- As there is no direct metric to measure the significance of the Yen carry trade as it relates to global markets or the effects of a Yen carry trade unwind, we aim to develop a proxy that sufficiently quantifies the effects of such as follows:
    - We will use USD/JPY (Dollar/Yen), fluctuations in Japanese Government Bond (JGB) yields, US-Japan base interest rate differentials and volatility in US equity markets via the Chicago Board Options Exchange's (CBOE) Volatility Index as a proxy for the unwinding of Yen carry trade.
        - The CBOE Volatility Index is a real-time index based upon S&P 500 index options and is a commonly used measure of the expected 30-day volatility for the S&P 500.
- Using this Yen carry trade unwind proxy we will explore the following:
    - Can we model a correlation between USD/JPY (Dollar/Yen), fluctuations in Japanese Government Bond (JGB) yields, US-Japan base interest rate differentials and volatility in US equity markets (CBOE VIX Index)?
    - Can a daily composite carry trade stress index whose components consist of the aforementioned Yen carry trade unwind proxy metrics with weighting determined by the loadings of principle component analysis (PCA) detect the Yen carry trade unwinding events of August 2024 and January 2026?
    - Will the composite carry trade stress index provide predictive modeling by showing elevated levels of stress (index values above baseline) prior to the Yen carry trade unwinding events of August 2024 and January 2026?


## Background and Prior Work

Big players in the global financial market (hedge funds, investment banks, mutual funds, etc.) borrow Yen for highly leveraged transactions due to the extremely low borrowing costs (~0.75% currently) since the stagnation of their economy began around 1990, known as the Yen carry trade.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) If the Yen increases in value due to forces in the Japanese bond market or domestic policy change then these investors are forced to liquidate their positions which can send shock waves through global markets known as the unwind of the Yen carry trade.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Japan recently elected a new Prime Minister, Sanae Takaichi, and she has called for significant tax cuts and an increase of government spending to combat persistent inflation despite Japan having the highest debt-to-GDP ratio in the world at 236.7 % in 2024.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) The Japanese government bond (JGB) market has reacted negatively to this plan with an increase in volatility that pushed the bond yields to all time highs, creating conditions for potential yen appreciation as Japanese institutions consider repatriating capital from foreign markets.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) <a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6)  This created the scenario where the Yen carry trade has started to unwind and the ripples of such were felt across the global markets. The ongoing situation in Japan has the potential to create a global financial crisis in the worst case scenario with some likening its potential ramifications to the 2008 global financial crisis.<a name="cite_note-7"></a> [<sup>7</sup>](#cite_ref-7) <a name="cite_note-8"></a> [<sup>8</sup>](#cite_ref-8)

In 2024 in the first week of August, the global and digital market saw a frantic shift in equities and assets being sold off in response to the unwinding of the Yen carry trade of Japan. The root cause of this being the Bank of Japan as it had increased the interest rate of the Japanese Yen from the range of 0% - 0.1% to 0.25% on July 31 of the same year. This sudden hike in the Japanese Yen’s interest rate in turn strengthened the appreciation of the yen over the US Dollar, which by early August had become a 10% increase in appreciation when previously it had been weakened against the US Dollar by 14% since the end of June 2024. Meanwhile on 2 August  2024, an employment report had shown the expected 179,000 jobs to be created in July was not met and only 114,000 jobs were reported. The combination of events resulted in a mass trade selloff, disrupting the market. As for 2026, by August the interest rate of the Japanese Yen had increased to 0.75%.

Some previous work related to Yen carry trade already exist. In 2009 IMF paper by Hattori and Shin tried to figure out if yen carry trades played a role in the 2008 crisis. The authors tracked how much yen foreign banks were moving out of Japan and compared it to the size of U.S. broker-dealer balance sheets over time—basically, when U.S. financial firms were growing their assets, the amount of yen leaving Japan increased proportionally, and when those firms shrank their balance sheets, yen flows reversed. They found a strong negative correlation (slope of -0.89) between changes in yen leaving Japan and domestic Japanese lending, showing that yen was being funneled abroad. Before the crisis, U.S. firms were borrowing cheap yen to invest in higher-yielding assets like mortgage-backed securities, but when the subprime mortgage market collapsed starting in 2007, it triggered a massive deleveraging across global financial markets. As U.S. institutions scrambled to sell assets and pay back their yen loans, this unwinding intensified the crisis.

Showing that a narrowing of the US-Japan base interest rate spread can lead to the Yen carry trade unwinding and that this leads to increased volatility in US equity markets.

## Dataset: Japanese Government Bond (JGB) 10-Year & 30-Year Benchmark Yields (2000–2026)
- Dataset Description:
    - This dataset focuses on the 10-Year (10Y) & 30-Year (30Y) Japanese Government Bond yields, measured in annualized percentages (%). During the 2000–2026 period, we see that there is a time of stability for around a decade before Japan goes into a zero and even negative interest rate for a few years. Additionally, we see that there is rapid yield increase between 2024 to 2026, which we are aiming to understand the implications of. One economic position to consider is that Japan's debt to GDP is one of the highest in the world. This could explain why the yield starts off steady and drops down to a negative value by 2016. By having a lower yield, Japan would avoid having to pay back high amounts of interest rates, avoiding default. By similar logic however, we can see that this rapidly increasing yield could definitely be a cause of concern.
- Major Concerns and Biases
    - A major concern to consider from this is from the idea explained above. If the Japanese government is trying to keep their economy afloat, they may be influencing the bond market in such a way that makes the data more synthetic. What this could mean is that if the economy begins to move out of the government's control, then the predictability for volatility could get difficult. 

```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
raw_path = 'data/00-raw/jgbcme_all.csv'
df = pd.read_csv(raw_path, skiprows=1)

# 2. Tidy & Filter (10Y / 30Y, 2000-2026)
df = df.replace('-', np.nan)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

df['10Y'] = pd.to_numeric(df['10Y'], errors='coerce')
df['30Y'] = pd.to_numeric(df['30Y'], errors='coerce')

mask = (df['Date'] >= '2000-01-01') & (df['Date'] <= '2026-12-31')
df_10y = df.loc[mask, ['Date', '10Y']].copy()
df_30y = df.loc[mask, ['Date', '30Y']].copy()

# 3. Size and Missingness Analysis
data_size_10y = len(df_10y)
na_count_10y = df_10y['10Y'].isna().sum()

data_size_30y = len(df_30y)
na_count_30y = df_30y['30Y'].isna().sum()

# 4. Outlier Flagging (4 SD rule)
# ---- 10Y ----
mean_val_10y = df_10y['10Y'].mean()
std_val_10y = df_10y['10Y'].std()
upper_limit_10y = mean_val_10y + (4 * std_val_10y)
lower_limit_10y = mean_val_10y - (4 * std_val_10y)

suspicious_entries_10y = df_10y[
    (df_10y['10Y'] > upper_limit_10y) | (df_10y['10Y'] < lower_limit_10y)
]
flag_count_10y = len(suspicious_entries_10y)

# ---- 30Y ----
mean_val_30y = df_30y['30Y'].mean()
std_val_30y = df_30y['30Y'].std()
upper_limit_30y = mean_val_30y + (4 * std_val_30y)
lower_limit_30y = mean_val_30y - (4 * std_val_30y)

suspicious_entries_30y = df_30y[
    (df_30y['30Y'] > upper_limit_30y) | (df_30y['30Y'] < lower_limit_30y)
]
flag_count_30y = len(suspicious_entries_30y)

# 5. Cleaning & Saving
os.makedirs('data/02-processed', exist_ok=True)

df_clean_10y = df_10y.dropna(subset=['10Y'])
df_clean_10y.to_csv('data/02-processed/jgb_10y_2000_2026_cleaned.csv', index=False)

df_clean_30y = df_30y.dropna(subset=['30Y'])
df_clean_30y.to_csv('data/02-processed/jgb_30y_2000_2026_cleaned.csv', index=False)

# 6. Visualization
sns.set_theme(style="whitegrid")

# ---- Plot A: 10Y only ----
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_clean_10y, x='Date', y='10Y', label='10Y Yield', color='royalblue', lw=1.5)

if flag_count_10y > 0:
    sns.scatterplot(
        data=suspicious_entries_10y, x='Date', y='10Y',
        color='red', label='Flagged (10Y)', zorder=5
    )

plt.axhline(upper_limit_10y, color='gray', linestyle='--', alpha=0.4, label='4 SD Threshold')
plt.axhline(lower_limit_10y, color='gray', linestyle='--', alpha=0.4)

plt.title('JGB 10-Year Yields (2000–2026)')
plt.ylabel('Yield (%)')
plt.xlabel('Year')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

print("--- Data Integrity Report (10Y) ---")
print(f"Total Observations: {data_size_10y}")
print(f"Missing (NaN) Data: {na_count_10y}")
print(f"Extreme Events Flagged: {flag_count_10y}")
print()

# ---- Plot B: 30Y only ----
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_clean_30y, x='Date', y='30Y', label='30Y Yield', color='purple', lw=1.5)

if flag_count_30y > 0:
    sns.scatterplot(
        data=suspicious_entries_30y, x='Date', y='30Y',
        color='red', label='Flagged (30Y)', zorder=5
    )

plt.axhline(upper_limit_30y, color='gray', linestyle='--', alpha=0.4, label='4 SD Threshold')
plt.axhline(lower_limit_30y, color='gray', linestyle='--', alpha=0.4)

plt.title('JGB 30-Year Yields (2000–2026)')
plt.ylabel('Yield (%)')
plt.xlabel('Year')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

print("--- Data Integrity Report (30Y) ---")
print(f"Total Observations: {data_size_30y}")
print(f"Missing (NaN) Data: {na_count_30y}")
print(f"Extreme Events Flagged: {flag_count_30y}")
print()
```

## Dataset: 10-Year U.S. Treasury yields (2000–2026)
- Dataset Description
    - This dataset focuses on the 10-Year U.S. Treasury yield, measured in annualized percentage. During the 2000–2026 period, the data illustrates a change from the relatively high-rate environment of the early years (above 6%) to a prolonged era of historic lows following the 2008 financial crisis. After a period of near-zero yields prompted by the 2020 COVID-19 pandemic, the dataset shows a dramatic economic shift starting in late 2022. Between 2024 and 2026, there is a sustained and rapid yield increase, with the rate climbing back toward the 4–5% range. This trend is largely driven by persistent inflationary concerns and the Federal Reserve’s aggressive policy changes toward monetary tightening, positioning the 10Y yield once again as a critical, high-carry benchmark.
- Major Concerns and Biases
    - A major concern regarding this dataset is the impact of "synthetic" demand created by the Federal Reserve’s policy of Quantitative Easing (QE). Because the central bank has historically intervened as a massive buyer of these bonds to suppress long-term rates, the yields during much of the 2010s and early 2020s may not purely reflect a  free-market equilibrium. Furthermore, the rapid volatility observed between 2024 and 2026 introduces potential bias where historical correlations may no longer hold true as the market adjusts to a post zero interest rate environment. There is also the risk that the data is influenced by traders, where market expectations are so heavily managed by policy announcements that the yield reflects anticipated government action rather than underlying economic health, potentially leading to increased "noise" in predictive models.

## Dataset: FEDERAL RESERVE BANK of ST.LOUIS - Japanese Yen to U.S. Dollar Spot Exchange Rate (2000-2026)
- Dataset description:
    - This dataset contains daily observations of the USD/JPY exchange rate from 2000 to 2026. The key metric, USD/JPY (DEXJPUS), is measured in units of Japanese yen per one U.S. dollar (JPY/USD). For example, a value of 110 indicates that one U.S. dollar can be exchanged for 110 Japanese yen. Because the rate is quoted as yen per dollar, higher values represent a weaker yen (more yen required to purchase one dollar), while lower values indicate a stronger yen. Over the sample period, USD/JPY has generally ranged between approximately 75 and 160, reflecting shifts in global risk sentiment and monetary policy conditions.
    - The exchange rate is a central variable for analyzing the yen carry trade. Since the strategy involves borrowing in yen and investing in higher-yielding assets abroad, movements in USD/JPY directly affect returns. Yen depreciation (rising USD/JPY) tends to support carry trade profitability, while yen appreciation (falling USD/JPY) can reduce returns and increase risk.
- Major Concerns and Biases
    - The dataset includes only the nominal daily USD/JPY exchange rate and does not incorporate interest rate differentials, inflation, or transaction costs, which are important for evaluating carry trade returns. Approximately 3.3% of observations are missing, primarily due to holidays and non-trading days.
    - Additionally, major macroeconomic events during the sample period, such as the 2008 Global Financial Crisis and others, may introduce structural shifts in the exchange rate over time.

```python
# 1. Load the dataset
raw_path = "data/00-raw/usd_jpy.csv"
df = pd.read_csv(raw_path)

# 2. Tidy & Filter (Yen per Dollar / 2000-2026)
df = df.rename(columns={
    "observation_date": "Date",
    "DEXJPUS": "YenPerDollar"
})

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["YenPerDollar"] = pd.to_numeric(df["YenPerDollar"], errors="coerce")

mask = (df["Date"] >= "2000-01-01") & (df["Date"] <= "2026-02-12")
df_JPUS = df.loc[mask, ["Date", "YenPerDollar"]].copy()

# 3. Size of data
data_size = len(df_JPUS)

# 4. Missing data info
na_count = df_JPUS.isna().any(axis=1).sum()

# I suspect that empty entries are not random and related to holidays/days off
# To test, I am checking to see how many New Years days have no entries
# If most, or all, New Years have no entries, it means that missing values are not random and are days off

df["month"] = df["Date"].dt.month
df["day"] = df["Date"].dt.day
new_year = df[(df["month"] == 1) & (df["day"] == 1)]
is_new_years_empty = new_year["YenPerDollar"].isna().mean()    # if == 1, it means all New Years entries have no value

    

# 5. Outlier Flagging
mean_val = df_JPUS["YenPerDollar"].mean()
std_val = df_JPUS["YenPerDollar"].std()
upper_limit = mean_val + (4 * std_val)    # anything beyond 4 std is extreme
lower_limit = mean_val - (4 * std_val)    # and will be flagged

suspicious_entries = df_JPUS[(df_JPUS["YenPerDollar"] > upper_limit) | (df_JPUS["YenPerDollar"] < lower_limit)]
flag_count = len(suspicious_entries)

# 6. Cleaning data
df_clean = df_JPUS.dropna()

# 7. Saving data
os.makedirs('data/02-processed', exist_ok=True)
df_clean.to_csv("data/02-processed/clean_exchange_rate_JPUS.csv", index=False)

# Seaborn Visualization
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

# Main plot for the dollar/yen exchange rate over time
sns.lineplot(data = df_clean, x="Date", y="YenPerDollar",label="Yen per Dollar")

# Flag sus on plot
if(flag_count > 0):
    sns.scatterplot(data=suspicious_entries, x="Date", y="YenPerDollar", color="red", label="Flagged Events", xorder=10)

# Add refernce lines for the SD threshold
plt.axhline(upper_limit, linestyle="--", alpha=0.4, label="4 SD Threshold")
plt.axhline(lower_limit, linestyle="--", alpha=0.4)

# Graph Labels/Style
plt.title("USD/JPY Exchange Rate (2000–2026)", fontsize=14)
plt.ylabel("Exchange Rate")
plt.xlabel("Year")
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()

# Data integrity report
print(f"--- Data Integrity Report ---")
print(f"Total Observations: {data_size}")
print(f"Missing (NaN) Data: {na_count}")
if(is_new_years_empty == 1):
    print("   Missing data is very unlikely to be random as all New Years days have no data, showcasing that it is likely holidays/days off")
print(f"Extreme Events Flagged: {flag_count}")
```

#### Correlation analysis between US and Japanese 10-year bond yields
- The objective of this exploratory analysis is to evaluate the historical relationship and spread between 10-Year U.S. Treasury yields and 10-Year Japanese Government Bond (JGB) yields from 2000 to 2026. To ensure a robust comparison, we integrated the two government datasets using an outer merge to preserve all trading days, employing forward-filling (ffill) to account for asynchronous national bank holidays. This methodology allows for a continuous calculation of the Yield Difference (Spread) without the data loss associated with cultural differences. To address the significant overplotting caused by over 6,500 daily observations, we utilized a 2D Kernel Density Estimate (KDE). This density heatmap effectively identifies the "modal" state of global interest rates, the purple core where yields most frequently clustered, while the overlayed regression line and Pearson correlation coefficient (r) quantify the strength of their linear association. In the second graph, we utilized a Time Series Comparison to track the differences of these yields over the given time period. Looking at these two graphs, we can extrapolate that the Japanese 10-Year yields and the US 10-Year yields influence one another especially during 2016 to 2023. Lastly, we would like to make note that since Japan has increased their interest rates, we see a sharp decline in the spread, which could potentially lead to a cause of concern for investors.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# 1. Explicitly load the fully wrangled data from data/02-processed
jp_clean_path = 'data/02-processed/jgb_10y_2000_2026_cleaned.csv'
us_clean_path = 'data/02-processed/us_10y_2000_2026_cleaned.csv'

df_jp = pd.read_csv(jp_clean_path, parse_dates=['Date'])
df_us = pd.read_csv(us_clean_path, parse_dates=['Date'])

# 2. Merge into ONE DataFrame for EDA
# Using outer merge + ffill to align differing national holidays
df_eda_1 = pd.merge(df_jp, df_us, on='Date', how='outer').sort_values('Date')

# Forward fill missing values caused by holidays, then drop remaining NaNs
df_eda_1[['10Y', '10Y_US']] = df_eda_1[['10Y', '10Y_US']].ffill()
df_eda_1 = df_eda_1.dropna()

# Add a column for the yield difference (US minus JP)
df_eda_1['Yield_Difference'] = df_eda_1['10Y_US'] - df_eda_1['10Y']

# Display a preview of the final merged EDA dataframe
print("Preview of the Merged EDA DataFrame:")
display(df_eda_1.head())

# 3. Calculate Pearson Correlation
correlation, p_value = stats.pearsonr(df_eda_1['10Y_US'], df_eda_1['10Y'])

# 4. Visualizations
sns.set_theme(style="whitegrid")

# --- Plot A: Correlation Scatter Plot ---
plt.figure(figsize=(10, 6))
sns.regplot(
    data=df_eda_1, 
    x='10Y_US', 
    y='10Y', 
    scatter_kws={'alpha': 0.3, 'color': 'purple', 's': 10}, 
    line_kws={'color': 'black', 'lw': 2, 'label': 'Trend Line'}
)

plt.title('Correlation between US and Japanese 10-Year Yields (2000-2026)', fontsize=14, pad=15)
plt.xlabel('US 10-Year Yield (%)', fontsize=12)
plt.ylabel('Japanese 10-Year Yield (%)', fontsize=12)

# Add a text box with the statistical findings
stats_text = f"Pearson r: {correlation:.4f}\nP-value: {p_value:.2e}"
plt.text(
    0.05, 0.95, stats_text, 
    transform=plt.gca().transAxes, 
    fontsize=12, verticalalignment='top', 
    bbox=dict(boxstyle='round', facecolor='white', alpha=0.9, edgecolor='gray')
)
plt.legend()
plt.tight_layout()
plt.show()


# --- Plot B: Time Series Comparison (Both Yields + The Spread) ---
plt.figure(figsize=(12, 6))

# Plot both yields
sns.lineplot(data=df_eda_1, x='Date', y='10Y_US', label='US 10Y Yield', color='forestgreen', lw=1.5)
sns.lineplot(data=df_eda_1, x='Date', y='10Y', label='Japan 10Y Yield', color='royalblue', lw=1.5)

# Plot the difference (Spread)
sns.lineplot(data=df_eda_1, x='Date', y='Yield_Difference', label='Spread (US - Japan)', color='purple', lw=1.5, linestyle='--')

# Add a reference line at zero for the spread
plt.axhline(0, color='black', linestyle=':', alpha=0.6)

plt.title('Comparative Timeline: US vs. Japanese 10-Year Yields & Spread', fontsize=14)
plt.ylabel('Yield / Spread (%)', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

# Print text interpretation for context
print("\n--- EDA Correlation Results ---")
print(f"Pearson Correlation Coefficient (r): {correlation:.4f}")
if correlation > 0.7:
    print("Interpretation: Strong positive linear correlation between US and JP yields.")
elif correlation > 0.3:
    print("Interpretation: Moderate positive linear correlation between US and JP yields.")
elif correlation < -0.7:
    print("Interpretation: Strong negative linear correlation between US and JP yields.")
else:
    print("Interpretation: Weak or no linear correlation between US and JP yields.")
```

This methodology allows the combined dataset to maintain continuity across the entire study period without losing observations due to differences in trading calendars. To visualize the relationship between the exchange rate and Japanese bond yields, scatter plots with fitted regression lines were used to evaluate the linear association between USD/JPY and both the 10-year and 30-year JGB yields, while Pearson correlation coefficients were calculated to quantify the strength of these relationships. ### Principle component analysis of USD/JPY, Japanese Government Bond yields, US-Japan base interest rate differentials and the CBOE Volatility Index
-  The objective if this exploratory data analysis (EDA) is to combine all of our variables of interest into one large DataFrame in order to better understand how they relate to each other and if there are any redundant variables.  This is carried out by generating general statistics of each variable, creating a correlation matrix and conducting principle component analysis (PCA).  The general statistics give us a high level overview of each variable and allow us to better understand what the data of each variable looks like side by side.  This side by side comparison is essential for gaining an intuition into the behavior of each variable and how their distributions differ.  The correlation matrix helps us to understand the linear relationship between each variable as it shows the Pearson's r-value for each possible combination of variables.  This allows insight into variables that are related to our target variable the CBOE VIX index and how changes in the Japanese financial markets affect US financial markets from a statistical viewpoint.  The PCA analysis identifies dimensions over which the most variance can be identified and the minimum number of components that need to be included in the Yen carry trade stress index in order to capture the underlying trends of the data without injecting noise.  This helps to show if we are adding unnecessary variables to our analysis and creates the weights used to develop the actual Yen carry trade stress index that this project first set out to create.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

# Create merged DataFrame combining all data
print('-' * 50 + ' Create merged DataFrame combining all data  ' + '-' * 50 )

# Load data from cleaned paths and df created in prior EDA
boj_rate_df = pd.read_csv('data/02-processed/BOJ_UOCR_2000_2026_cleaned.csv')
us_rate_df = pd.read_csv('data/02-processed/FRED_FFER_2000_2026_cleaned.csv')
vix_df = pd.read_csv('data/02-processed/vix_clean.csv')
fx_df = pd.read_csv('data/02-processed/clean_exchange_rate_JPUS.csv')
jgb_30y_df = pd.read_csv('data/02-processed/jgb_30y_2000_2026_cleaned.csv')
us_jpy_yield_df = df_eda_1

# Create unified date column naming
boj_rate_df.rename(columns={'Series code': 'DATE'}, inplace=True)
us_rate_df.rename(columns={'observation_date': 'DATE'}, inplace=True)
vix_df.rename(columns={'Date': 'DATE'}, inplace=True)
fx_df.rename(columns={'Date': 'DATE'}, inplace=True)
us_jpy_yield_df.rename(columns={'Date': 'DATE'}, inplace=True)
jgb_30y_df.rename(columns={'Date': 'DATE'}, inplace=True)

# Convert date columns to datetime if they aren't already
boj_rate_df['DATE'] = pd.to_datetime(boj_rate_df['DATE'])
us_rate_df['DATE'] = pd.to_datetime(us_rate_df['DATE'])
vix_df['DATE'] = pd.to_datetime(vix_df['DATE'])
fx_df['DATE'] = pd.to_datetime(fx_df['DATE'])
jgb_30y_df['DATE'] = pd.to_datetime(jgb_30y_df['DATE'])

# Rename columns to better reflect values
boj_rate_df.rename(columns={'FM01\'STRDCLUCON': 'BOJ_RATE'}, inplace=True)
us_rate_df.rename(columns={'RIFSPFFNB': 'US_RATE'}, inplace=True)
vix_df.rename(columns={'Close': 'VIX_CLOSE'}, inplace=True)
fx_df.rename(columns={'YenPerDollar': 'JPY_VS_USD'}, inplace=True)
us_jpy_yield_df.rename(columns={'10Y': 'JP_10Y'}, inplace=True)
us_jpy_yield_df.rename(columns={'10Y_US': 'US_10Y'}, inplace=True)
us_jpy_yield_df.rename(columns={'Yield_Difference': 'US_JP_10Y_DIF'}, inplace=True)
jgb_30y_df.rename(columns={'30Y': 'JP_30Y'}, inplace=True)

# Only consider VIX close for combined data analysis
vix_df.drop(labels=['High', 'Low', 'Open', 'Volume'], axis=1, inplace=True)

# Merge into one DataFrame for EDA
pca_df = pd.merge(boj_rate_df, us_rate_df, on=['DATE'], how='outer')
pca_df = pd.merge(pca_df, vix_df, on=['DATE'], how='outer')
pca_df = pd.merge(pca_df, fx_df, on=['DATE'], how='outer')
pca_df = pd.merge(pca_df, us_jpy_yield_df, on=['DATE'], how='outer')
pca_df = pd.merge(pca_df, jgb_30y_df, on=['DATE'], how='outer')

# Forward fill missing values caused by holidays, then drop remaining NaNs
pca_df= pca_df.ffill()
pca_df.dropna(inplace=True)

# Add a column for the US-Japan interest rate spread (US minus JP)
pca_df['US_JP_RATE_DIF'] = pca_df['US_RATE'] - pca_df['BOJ_RATE']

# Display a preview of the merged dataframe
display(pca_df.head())
```

- Combing all of our variables of interest into one large DataFrame allows for ease of analysis and is the first step towards creating the Yen carry trade stress index.

```python
# EDA on combined data
print('-' * 62 + ' EDA on combined data  ' + '-' * 62)
pca_numeric = pca_df[['BOJ_RATE', 'US_RATE', 'VIX_CLOSE', 'JPY_VS_USD', 'JP_10Y', 'JP_30Y', 'US_10Y', 'US_JP_10Y_DIF', 'US_JP_RATE_DIF']]
display(pca_numeric.describe())

# Create correlation matrix
plt.subplots(figsize=(15, 10))
pca_corr = pca_numeric.corr()
corr_plot = sns.heatmap(pca_corr, annot=True).set_title('Correlation matrix of variables used to create carry trade stress index')
display(corr_plot)

# PCA analysis
pca = Pipeline([('scaler', StandardScaler()), ('pca', PCA(n_components=9))])
pca.fit(pca_numeric)

explained_variance = pca[-1].explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

plt.figure(figsize=(15,10))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.axhline(y=0.9, color='r', linestyle='--', label='90% Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Number of Principal Components')
plt.legend()
plt.grid(True)
plt.show()
```

- Numerical description of combined data (mean/std/min/25%/50%/75%/max)
    - The maximum spread between US and Japanese base interest rate (`US_JP_RATE_DIF`) is 7%!
        - In terms of interest rates thats a massive spread and shows why traders actively engage in the Yen carry trade as borrowing Yen is significantly cheaper on average than USD with a mean value of around 2%.
- Correlation matrix 
    - Of the variables we plan to use in our carry trade stress index, it can be observed that there is a wide range of values.
    - Interestingly, the spread between US and Japanese base interest rate (`US_JP_RATE_DIF`), Japanese Yen per US dollar (`JPY_VS_USD`), and  10-year Japanese bond yields (`JP_10Y`) show somewhat significant correlation. 
    - Also the daily closing value of the CBOE VIX index (`VIX_CLOSE`) has quite low correlations with all other variables which could prove to be problematic for our initial hypothesis as we predicted there would be a significant correlation between its value and variables related to the Yen carry trade.
- Principle component analysis (PCA)
    - The PCA analysis shows that over 90% of the variance can be explained by only 4 components!
        - This implies that many of our variables are highly correlated and that we may be injecting noise into the system if we choose to include all of them.
            - This makes sense as the PCA analysis includes variables such as the spread between US and Japanese base interest rate (`US_JP_RATE_DIF`) and the components of the calculations i.e. US and Japanese base interest rates (`BOJ_RATE` and `US_RATE`).
                - Depending on the results of the initial carry trade stress index, we may choose to remove the components of variables used to compute some other variable to avoid injecting noise into the system i.e `BOJ_RATE`, `US_RATE`, `JP_10YR`, `US_10YR`

```python
from sklearn.preprocessing import MinMaxScaler

# Yen Carry Trade Stress Index
print('-' * 55 + ' Yen Carry Trade Stress Index  ' + '-' * 55)

# Loadings of PCA analysis
loadings = pca['pca'].components_
loadings_df = pd.DataFrame(pca['pca'].components_, columns=pca_numeric.columns, index=[f'PC{i+1}' for i in range(loadings.shape[0])])
display(loadings_df)

# Create stress index DataFrame by taking the dot product between numeric variables and PCA loadings
stress_df = pca_df[['DATE']]
pca_weights = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9']
for component in pca_weights:
    stress_df[f'STRESS_{component}'] = pca_numeric.dot(loadings_df.loc[component])

# Normalize stress index to 0-100
scaler = MinMaxScaler(feature_range=(0, 100))
for i in range(1, 10):
    col = f'STRESS_PC{i}'
    stress_df[col] = scaler.fit_transform(stress_df[[col]])
display(stress_df.head())

# Graph the first 4 stress indexes vs. the VIX
for i in range(1, 5):
    fig, ax1 = plt.subplots(figsize=(15, 8))
    sns.lineplot(data=stress_df, x='DATE', y=f'STRESS_PC{i}', ax=ax1, label='Stress Index')
    ax1.set_ylabel(f'PC{i} Stress Index')
    ax2 = ax1.twinx()
    sns.lineplot(data=pca_df, x='DATE', y='VIX_CLOSE', ax=ax2, color='green', alpha=0.3, label='VIX')
    ax2.set_ylabel('VIX')
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    ax2.grid(False)
    plt.title(f'PC{i} Carry Trade Stress Index vs VIX')
    plt.show()
```

- The above Yen carry stress indexes are derived by taking the element wise product of the raw principle component loadings against their associated variable and summing the result (dot product).  This value is then scaled from 0-100 to generate a more intuitive index that can be compared to more traditional indexes like the CBOE VIX.
- When comparing the Yen carry trade stress index to the VIX index and looking at the raw weights we can start to get an idea of the trends that each component is capturing.
    - Principle component one (PC1), two (PC2), and four (PC4) appear to barely resemble the VIX, but likely are capturing trends in both US and Japanese financial markets with PC1 and PC4 producing very similar looking timeseries.
    - Principle component three (PC3) appears to capture the underlying trend of the VIX, looking almost identical through the plotted range.
    - Since over 90% of the variance can be explained by the first 4 components, we're going to only use those for the creation of the final stress index in the next step.

```python
# Create a blended stress index using the first 4 PCA stress weights
# Use explained variance ratios as weights for PCA stress components
n_components = 4
variances = pca['pca'].explained_variance_ratio_[:n_components]
weights = variances / variances.sum()

combined_loadings = loadings_df.iloc[:n_components]
blended_loadings = weights @ combined_loadings 
stress_df['STRESS_BLENDED'] = pca_numeric.dot(blended_loadings)
stress_df['STRESS_BLENDED'] = scaler.fit_transform(stress_df[['STRESS_BLENDED']])
display(stress_df.head())

# Graph blended stress index vs. the VIX
fig, ax1 = plt.subplots(figsize=(15, 8))
sns.lineplot(data=stress_df, x='DATE', y='STRESS_BLENDED', ax=ax1, label='Stress Index')
ax1.set_ylabel('Blended PC Stress Index')
ax2 = ax1.twinx()
sns.lineplot(data=pca_df, x='DATE', y='VIX_CLOSE', ax=ax2, color='green', alpha=0.3, label='VIX')
ax2.set_ylabel('VIX')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
ax2.grid(False)
plt.title('Blended PC Carry Trade Stress Index vs VIX')
plt.show()

# Graph blended stress index for for 01-2018 to 02-2026
mask = (stress_df['DATE'] >= '2018-02-15') & (stress_df['DATE'] <= '2026-02-15')
stress_2018_to_2026 = stress_df[mask]
fig, ax1 = plt.subplots(figsize=(15, 8))

sns.lineplot(data=stress_2018_to_2026, x='DATE', y='STRESS_BLENDED', ax=ax1, label='Stress Index')
ax1.fill_between(stress_2018_to_2026['DATE'], 19, stress_2018_to_2026['STRESS_BLENDED'], alpha=0.15, color='#1f77b4')

ax1.set_ylabel('Blended PC Stress Index')
ax1.legend(loc='upper right')
plt.title('Blended PC Carry Trade Stress Index 02-2018 to 02-2026')
plt.show()

# Graph blended stress index for 2024 only
stress_2024 = stress_df[stress_df['DATE'].dt.year == 2024]
fig, ax1 = plt.subplots(figsize=(15, 8))

sns.lineplot(data=stress_2024, x='DATE', y='STRESS_BLENDED', ax=ax1, label='Stress Index', linewidth=2)
ax1.fill_between(stress_2024['DATE'], 59, stress_2024['STRESS_BLENDED'], alpha=0.15, color='#1f77b4')

ax1.set_ylabel('Blended PC Stress Index')
ax1.legend(loc='upper right')
plt.title('Blended PC Carry Trade Stress Index 2024')
plt.show()

# Graph blended stress index for 02-2025 to 02-2026
mask = (stress_df['DATE'] >= '2025-02-15') & (stress_df['DATE'] <= '2026-02-15')
stress_2026 = stress_df[mask]
fig, ax1 = plt.subplots(figsize=(15, 8))

sns.lineplot(data=stress_2026, x='DATE', y='STRESS_BLENDED', ax=ax1, label='Stress Index', linewidth=2)
ax1.fill_between(stress_2026['DATE'], 65, stress_2026['STRESS_BLENDED'], alpha=0.15, color='#1f77b4')

ax1.set_ylabel('Blended PC Stress Index')
ax1.legend(loc='upper right')
plt.title('Blended PC Carry Trade Stress Index 02-2025 to 02-2026')
plt.show()
```

- This blended stress index takes the amount of variance that can be explained by a particular principle component and uses that as the components weighting.  A blended loading is then derived by computing a matrix multiplication between the weights and the loadings of each of the four components.  The dot product of this blended loading is then taken between the associated variables and scaled from 0-100 as in the previous stress indexes.
- It can be observed that the spikes from increased volatility in US equities markets are reflected in the stress index in addition to a trend of continually increasing stress since 2022.
- As was predicted in the hypothesis there is a significant spike around August of 2024 with the unwinding of the Yen carry trade due to the Bank of Japan raising interest rates.
- Interestingly the stress index does spike in early 2026, but not in a meaningful way above the local maximums.  This is in disagreement with our initial hypothesis and it appears that we were unable to capture the forces at play for the Yen carry trade unwinding event earlier this year.
- One thing to note on about the graph of the carry trade stress index from 02-2025 to 02-2026 is that there is a massive spike around April of 2025.  This correlates perfected with president Trump liberation Day tariffs that imposed wide spread tariffs on global trade partners and sent financial markets into a short lived panic.  This shows that we likely are capturing general trends in global financial markets and we're not able to isolate the trends that solely capture movements in Yen carry trades.  However, this is to be expected as financial data is extremely noisy and isolating one specific trade from any kind of outside noise is not possible.

### D. Modeling
 - [x] **D.1 Proxy discrimination**: Have we ensured that the model does not rely on variables or proxies for variables that are unfairly discriminatory?

 - [x] **D.2 Fairness across groups**: Have we tested model results for fairness with respect to different affected groups (e.g., tested for disparate error rates)?

Our model focuses on macro-level indicators (JGB yields and global volatility); however, "unwind" events have disparate impacts on different market participants. High-frequency and institutional traders often have faster access to exit liquidity, while retail investors or smaller funds may bear a disproportionate burden of the subsequent market crash. To address this, we will compare our model's "stress alerts" against historical data for both large-cap and smaller-cap indices to evaluate if our index identifies risks equally well for different sectors.

 - [x] **D.3 Metric selection**: Have we considered the effects of optimizing for our defined metrics and considered additional metrics?

While the VIX serves as our primary validation metric for global equity volatility, we recognize that it may not fully capture the nuances of the Yen carry trade. To address this, we have implemented a multi-metric optimization strategy that incorporates the  volatility of Japanese markets. Specifically, we are calculating realized volatility for the USD/JPY exchange rate and yield fluctuations for 10-year and 30-year Japanese Government Bonds (JGBs). By triangulating these mathematical measures, we ensure our stress index accounts for "funding currency" liquidity risks and interest rate differentials, rather than acting as a mere proxy for U.S. market sentiment.

 - [x] **D.4 Explainability**: Can we explain in understandable terms a decision the model made in cases where a justification is needed?

Our model utilizes Principal Component Analysis (PCA) to ensure the "Stress Index" remains interpretable. By analyzing the factor loadings of each principal component, we can explicitly quantify which financial variables, such as JGB yield shifts or JPY/USD volatility, most heavily influenced the model’s output during a specific period. This provides a transparent rationale for any "stress" signals generated. While we acknowledge that financial markets are influenced by various erratic variables, we remove complexity by focusing our explanations on these high-impact macroeconomic drivers, effectively translating high-dimensional data into understandable market narratives.

 - [x] **D.5 Communicate limitations**: Have we communicated the shortcomings, limitations, and biases of the model to relevant stakeholders in ways that can be generally understood?

We will explicitly define our model as a retrospective stress index intended for educational and academic purposes, rather than a framework for active financial trading or professional investment advice. We acknowledge the human tendency for pattern-seeking and the risk of users applying these findings to future market movements. To avoid this, our conclusion will explicity state a disclaimer emphasizing that our analysis provides a high-level, "macro-scale" view of the bilateral relationship between Japan and the United States. We will clarify that the model lacks the granularity required for high-frequency or short-term trading and should not be used as a primary decision-making tool in live markets.
