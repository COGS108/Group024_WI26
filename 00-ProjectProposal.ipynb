## Research Question

What is the significance of the yen carry trade as it relates to global markets?
- Can a composite carry trade stress index derived from Japanese Government Bonds (JGB) yield volatility and US-Japan interest rate differentials reliably quantify systemic risk and detect the Yen carry trade unwinding events of August 2024 and January 2026?

## Background and Prior Work

Big players in the global financial market (hedge funds, investment banks, mutual funds, etc.) borrow Yen for highly leveraged transactions due to the extremely low borrowing costs (~0.75% currently) since the stagnation of their economy began around 1990, known as the Yen carry trade.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) If the Yen increases in value due to forces in the Japanese bond market or domestic policy change then these investors are forced to liquidate their positions which can send shock waves through global markets known as the unwind of the Yen carry trade.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Japan recently elected a new Prime Minister, Sanae Takaichi, and she has called for significant tax cuts and an increase of government spending to combat persistent inflation despite Japan having the highest debt-to-GDP ratio in the world at 236.7 % in 2024.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) The Japanese government bond (JGB) market has reacted negatively to this plan with an increase in volatility that pushed the bond yields to all time highs, creating conditions for potential yen appreciation as Japanese institutions consider repatriating capital from foreign markets.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) <a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6)  This created the scenario where the Yen carry trade has started to unwind and the ripples of such were felt across the global markets. The ongoing situation in Japan has the potential to create a global financial crisis in the worst case scenario with some likening its potential ramifications to the 2008 global financial crisis.<a name="cite_note-7"></a> [<sup>7</sup>](#cite_ref-7) <a name="cite_note-8"></a> [<sup>8</sup>](#cite_ref-8)

## Hypothesis


A principle component analysis (PCA) based index constructed from normalized JGB market indicators will exhibit statistically significant spikes preceding or concurrent with global volatility surges (VIX) during the August 2024 and January 2026 events. 

## Data


1. The ideal dataset for this project would be a multi-source time-series dataset that captures indicators of yen carry trade stress in Japanese markets alongside measures of volatility in US equity markets. Since no single dataset directly measures carry trade activity, we will construct a composite dataset by merging multiple publicly availabl sources.
    1. Key variables: USD/JPY (Dollar/Yen), Japenese Government Bond (JGB) yields and yeild volatility, US-Japan interest rate differentials, global market volatility metrics (VIX index), and date/time index for merging datasets
    2. Ideally, the dataset will contain daily/weekly observations over the span of multiple years(e.g., 2010-2026).
    3. Financial market data is collected automatically by exchanges, central banks, and economic data aggregators rather than surveys or expirements.
    4. The data can be orginazed in a time-indexed tabular format, where each row represents a date and each column represents a financial variable. After getting data from different sources, the datasets would be merged using the timestamp and stored as cleaned pandas Dataframes for analysis.

2. These are the data sources we can use:

    1. FRED (Federal Reserve Economic Data)
        1. URL: "https://fred.stlouisfed.org/"
        2. FRED provides downloadable CSV time-series datasets from central banks and financial institutions without the need of permission or application. The variables that are important include: Japanese yen exchange rate(USD/JPY), Japanese interest rate and government bond yields, VIX volatility index(global market risk indicator).
  
    2. Trading Economics
        1. URL: "https://www.mof.go.jp/english/policy/jgbs/reference/interest_rate/index.htm"
        2. Historical Japanese Government Bond interest rate data is publicy available and downloadable as CSV files. The variables that are important include: JGB yield levels across maturities and yield chanages or volatility.

    3. BOJ (Bank of Japan) Time-Series Data
        1. URL: "https://www.stat-search.boj.or.jp/index_en.html"
        2. This site is another that offers downloadable CSV files. BOJ also publishes official macroeconomic and financial statistics updated multiple times daily. Important variables include: policy interest rates, monetary policy indicators, and government bond market metrics.
